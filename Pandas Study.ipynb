### Zamjena pomoću mean, median i mode

Često se nepostojeće vrijednosti u stupcu nadomještaju sa nekim statističkim vrijednostima

#### MEAN (srednja vrijednost) - prosjek
#### MEDIAN (medijan) - srednji član kada se vrijednosti poredaju uzlazno (ili prosjek dva srednja člana za parni broj članova)
#### MODE (mod) - vrijednost koja se pojavljuje najviše puta; ako se sve pojavljuju isti broj puta, onda je svaka mod

```python
data4=pd.read_csv("MetObjects.csv", engine='python')
data4
```

```python
data4.info()
```

```python
mean_cals = data4["Calories"].mean()
mean_cals
```
